## Величины


Величина — это все, что можно измерить линейкой, причем необязательно буквально: рост, вес, температура, стоимость, время и другие показатели легко выразить числами, которые поймут люди. Проще говоря, все, что можно выразить цифрами или набором цифр, называют величинами.

Цвет, форма и эмоции — не величины, потому что их нельзя выразить цифрами. С величинами всегда понятно, что больше, а что меньше. Красный не больше и не меньше, чем синий, это просто два разных цвета. А радость и печаль — просто разные эмоции, и считать тут особо нечего.

## Что такое переменная

В этом мире нет ничего постоянного, однако  что-то условно меняется, а что-то нет. Температура на улице, зарплата сотрудника, время в пути и так далее — все это называют переменными.

Например, продолжительность суток, температура кипения воды в чайнике, рост человека с 25 до 30 лет если и меняются, то незначительно. Такое называют константами.

Оба типа величин используют в программировании, однако переменные — намного чаще, особенно в Питоне.

## Переменные в Python

В Python общение с памятью происходит немного иначе, потому что там вы редко работаете с реальным физическим адресом переменной. Разработчики дают переменным имена.

Например, вы говорите Python, что нужна переменная ```weekday```, чтобы каждый раз не писать день недели вручную.

````python
import datetime

weekday = datetime.datetime.today().weekday()

print('day number:', weekday)

weekdays_list = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

print('day name:  ', weekdays_list[weekday])
````

Сейчас не очень понятно, что тут происходит, но структура в коде уже читается.

Разработчики выбирают Python из-за огромного количества дополнительных инструментов, которые можно добавлять к основному арсеналу.

```import datetime``` — добавление группы функций, связанных с датами и временем. На экране ничего не появится, однако запуск этого кода позволит использовать ящик с инструментами под называнием ```datetime```.




````python
import datetime
````


```datetime.datetime.today().weekday()``` — загадочная функция, которая при запуске дает номер сегодняшнего дня. Важно: отсчет идет от 0, поэтому нулем будет понедельник. При наборе кода была пятница, поэтому результат 4.

````python
datetime.datetime.today().weekday()
````

Результат работы этой функции мы аккуратно складываем в переменную ```weekday```. Тут при запуске мы снова ничего не увидим, но в памяти кое-что изменится — появится ящик ```weekday```.

````python
weekday = datetime.datetime.today().weekday()
````

## Отличия Python от Excel

В Excel вы прямо записывали что-нибудь в конкретную ячейку.

Тут в D3 вписывают текст "week day" практически напрямую в память.

Python сам выбирает в памяти какую-нибудь свободную ячейку, запоминает, что мы решили называть ее ```weekday```, и складывает в нее то, что мы попросим. В данном случае результат работы какой-то функции.

```weekday = datetime.datetime.today().weekday()``` означает следующее:

«Python, запусти функцию ```datetime.datetime.today().weekday()```. Результат работы этой функции (число) сложи в какое-нибудь свободное в памяти место. Это свободное место будем называть ```weekday```».

Что именно это за место, знать не нужно. По аналогии с Excel это могла быть ячейка D5 или C8, но Python сам решает, какую ячейку использовать. Для нас это место называется ```weekday```.

Теперь, когда результат работы функции сохранен, можно его переиспользовать много раз.

Сначала просто выводим то, что было сложено в переменную ```weekday```.

2.3.png

````python
print('day number:', weekday)
````

Питон, выведи на экран текст "day number:" и то, что лежит в переменной ```weekday```. Мы видим текст ```day number: 4```. То есть в переменной ```weekday``` дейсвительно лежит число 4 (потому на момент написания кода была пятница).

Далее у нас произойдет магия (для новичков Python), но суть будет такая же, как в Excel. Тут вместо номера дня недели мы хотим получить названия этого дня недели.

2.4.png

Если запустить этот код, то мы увидим не число, которое сложили в переменную ```weekday```, а соответствующее название дня недели.

````python
weekdays_list = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

print('day name:  ', weekdays_list[weekday])
````

Выходит ```fri```.

Получается, переменная — это область памяти, в которую вы вписываете информацию, а потом можете многократно получать доступ к этой области памяти и ее содержимому по заданному имени.

## Как назвать переменную

Переменные — они как дети. Имена для своих детей мы придумываем сами, однако хаотично называть людей не принято и есть правила. Причем у правил этих два уровня.
* Первый уровень — это государственный закон (правила Python), согласно которому имя не может быть случайным набором латинских или китайских символов. Это должно быть что-то читаемое и написанное на кириллице, если речь о России.
* Второй уровень — это закон человеческий (программистский). Никто не запрещает вам называть ребенка «Рэмбо», однако у других людей (программистов или правообладателей) могут возникнуть к вам вопросы.
Начнем с первого уровня законов.

### Законы именования Питона



1.   Название переменной должно начинаться с латинского символа или подчеркивания.
2.   Название переменной не может начинаться с цифры.
3.   Название переменной может содержать только строчные и заглавные буквы латинского алфавита, цифры и подчеркивание (А–z, 0–9 и _).
4.   Регистр имеет значение (Week и week — это разные имена).
5.   Есть [список запрещенных имен](https://www.w3schools.com/python/python_ref_keywords.asp), зарезервированных самим Python (например: def, and, True, if и так далее).

### Закон человеческий

Именование переменных бесконечно обсуждают в мире программирования, но со временем у сообщества появились более-менее общепринятые стандарты. Главная идея, вокруг которой крутятся все правила, — удобство.

Эти законы нарушать можно, и Python разрешит вам, но коллеги не одобрят.

Главное правило — избегать сокращений. Бывают исключения, но очень редкие.

Вот так делать можно:

````python
weekday = datetime.datetime.today().weekday()
````

Название явно показывает, зачем нужна переменная. Ровно так же работали имена людей в древности, и так они до сих пор работают в различных религиозных или околорелигиозных культурах. Человек получает имя согласно своим качествам или функции. Александр — воин, Мария — госпожа, Сократ — разумный, Платон — широкоплечий.

Вот так делать плохо:

````python
weekd = datetime.datetime.today().weekday()
wday = datetime.datetime.today().weekday()
wd = datetime.datetime.today().weekday()
x = datetime.datetime.today().weekday()
````

И плохо в первую очередь для вас. Потому что, когда вы увидите эти строчки, появятся вопросы и придется искать, что вы вообще складывали в переменную. Следующий код будет работать, но читать его сложно.

````python
print('day name:', weekdays_list[weekd])
print('day name:', weekdays_list[wday])
print('day name:', weekdays_list[wd])
print('day name:', weekdays_list[x])
````

Вот так сразу ясно, что происходит:

````python
print('day name:  ', weekdays_list[weekday])
````

Встречаются общепринятые сокращения. Если вы собрались построить график функции, то допустимы переменные с именами ```x``` и ```y```.

Но слишком подробно расписывать назначение переменной тоже не нужно, иначе код будет очень длинным. Понимание «золотой серидины» придет с опытом.
