# Основные типы данных Python


Информация, или данные, — это загадочная концепция, но в сущности все просто. Данные — то, что помогает узнать что-то о чем-то. Каждый человеческий язык накладывает свои ограничения на то, что можно и нельзя описать, и языки программирования не исключение. В Python есть ограниченный набор типов того, что можно записывать в память и чем можно описывать что-нибудь.

## Какие бывают числа

О числе можно думать как о забавном наборе каракулей, предназначенном для того, чтобы описать какое-то количество или величину. Идея очень древняя.

image.png

В математике примерно 5 базовых типов чисел, хотя на самом деле их намного больше. В программировании все проще, базовых типов всего 2.

## Integer

Integer — это целые числа. Целыми числами можно считать овец, людей, деревья и вообще любые неделимые вещи. Овцу, конечно, поделить можно... но она скорее всего перестанет быть полноценной овцой. Как и дерево. Такие величины называются целыми.

2.6.png

Если просто ввести в консоль 3 и запустить, то это уже целое число.

```python
3
```

Ничего интересного не произойдет, но надо же с чего-то начинать! Запишите число в переменную, а потом выведите:

```python
sheep_number = 3
print('число овец: ', sheep_number)
```

Вот вы и описали какое-то явление природы. Теперь эту переменную можно переиспользовать. Например, увеличить число в переменной на 1.

```python
sheep_number += 1
print('число овец: ', sheep_number)
```

Скоро вы узнаете, как это работает, а пока просто потренеруемся записывать информацию и выводить.
